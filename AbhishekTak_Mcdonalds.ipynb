# PCA Analysis

```python

from pca import pca
pca_model = pca(n_components = 11)
results = pca_model.fit_transform(Matrix.to_numpy(),col_labels = Matrix.columns)
```

```python
print(pca_model.results['explained_var'].round(4))
```

```python
print(pca_model.results['variance_ratio'].round(4))
fig, ax = pca_model.plot()
```

```python
pca_matrix = pd.DataFrame(results['loadings'])
display(pca_matrix.T)
```

# Step 5 (Extracting segments)

```python
from sklearn.cluster import KMeans
```

```python
inertias = []

for i in range(2,9):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(Matrix)
    inertias.append(kmeans.inertia_)

plt.plot(range(2,9), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
k_means = KMeans(n_clusters=4)
k_means.fit(Matrix)
```

```python
label = k_means.fit_predict(Matrix)
```
